```python
import pandas as pd

# Load original data
movies = pd.read_csv("ml-25m/movies.csv")
ratings = pd.read_csv("ml-25m/ratings.csv")
genome_scores = pd.read_csv("ml-25m/genome-scores.csv")
links = pd.read_csv("ml-25m/links.csv")
# Select top 2k most-rated movies
top_movies = (
    ratings.groupby("movieId")
    .size()
    .sort_values(ascending=False)
    .head(2000)
    .index
    
)

# Filter datasets
movies_2k = movies[movies.movieId.isin(top_movies)]
ratings_2k = ratings[ratings.movieId.isin(top_movies)]
genome_2k = genome_scores[genome_scores.movieId.isin(top_movies)]
links_2k = links[links.movieId.isin(top_movies)]
# Save reduced data
movies_2k.to_csv("data/movies_2k.csv", index=False)
ratings_2k.to_csv("data/ratings_2k.csv", index=False)
genome_2k.to_csv("data/genome_scores_2k.csv", index=False)
links_2k.to_csv("data/links_2k.csv", index=False)
print("✅ Mini dataset created")
```
